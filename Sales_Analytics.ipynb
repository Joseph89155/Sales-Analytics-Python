### Load and Previewing the Data
Dataset - [link text](https://www.kaggle.com/datasets/iamsouravbanerjee/customer-shopping-trends-dataset/)

```python
# Loading relevant libraries and the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data
sales = pd.read_csv('shopping_trends_updated.csv')
sales.head()
```

```python
# Previewing the bottom of the dataset
sales.tail()
```

```python
# Checking on dataset shape
sales.shape
```

### Cleaning the data

```python
# Checking on the datatype
sales.info()
```

```python
# Check for missing values
sales.isnull().sum()
```

```python
# Checking for duplicates
sales.duplicated().sum()
```

### Data Exploration and Visualization

```python
# Summary of the data
sales.describe()
```

```python
# Q1. Print out the most popular payment method.
sales.groupby('Payment Method')['Payment Method'].count().sort_values(ascending=False)
```

```python
# Visualization
import seaborn as sb
base_color = sns.color_palette()[2]
method = sales['Payment Method'].value_counts().index
sb.countplot(data = sales, x = 'Payment Method', color = base_color, order = method)
```

```python
# Q2. Print out the most popular item purchased
sales.groupby('Item Purchased')['Item Purchased'].count().sort_values(ascending=False)
```

```python
# Visualization
import seaborn as sb
base_color = sns.color_palette()[2]
method = sales['Item Purchased'].value_counts().index
sb.countplot(data = sales, y = 'Item Purchased', color = base_color, order = method)
```

```python
# Q3, print out the popular item purchased by females
sales[sales['Gender'] == 'Female'].groupby('Item Purchased')['Item Purchased'].count().sort_values(ascending=False).head(1)
```

```python
#Q4. Visualize the locations that drives majority of the sales
location = sales['Location'].value_counts().head(10).index
sb.countplot(data = sales, y = 'Location', color = base_color, order = location)
```

```python
#Q5. On average how much do customers spend
mean_purchases = round(sales['Purchase Amount (USD)'].mean(),2)
median_purchases = round(sales['Purchase Amount (USD)'].median(),2)

print(f'The mean purchase amount is {mean_purchases} and the median purchase amount is {median_purchases}')
```
